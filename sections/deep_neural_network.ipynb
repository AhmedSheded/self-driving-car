```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
np.random.seed(2222)
```

```python
points = 500
X, y =  datasets.make_circles(n_samples=points, random_state=2222, noise=0.15, factor=0.2)
```

```python
plt.scatter(X[y==0, 0], X[y==0, 1])
plt.scatter(X[y==1, 0], X[y==1, 1])
plt.show()
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
model = Sequential()
model.add(Dense(4, input_shape=(2,), activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
model.compile(Adam(learning_rate=0.01), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(x=X, y=y, verbose=1, batch_size=20, epochs=100, shuffle=True)
```

```python
plt.plot(history.history['accuracy'])
```

```python
def plot_decision_boundary(X, y, model):
    x_span = np.linspace(min(X[:, 0])-0.25, max(X[:, 0])+0.25)
    y_span = np.linspace(min(X[:, 1])-0.25, max(X[:, 1])+0.25)
    xx, yy = np.meshgrid(x_span, y_span)
    xx_, yy_ = xx.ravel(), yy.ravel()
    grid=np.c_[xx_, yy_]
    pred = model.predict(grid)
    z= pred.reshape(xx.shape)
    plt.contourf(xx, yy, z)
```

```python
plot_decision_boundary(X, y, model)
plt.scatter(X[:points,0], X[:points,1])
plt.scatter(X[points:,0], X[points:,1])
```
